```python
for char in string.punctuation:
  print(char, end=" ")
translate_table = dict((ord(char), None) for char in string.punctuation)
```

```python
import re
data_eng = []
lang_eng = []
for line in df_eng:
  if len(line) !=0:
    line = line.lower()
    line = re.sub(r"\d+", "", line)
    line = line.translate(translate_table)
    data_eng.append(line)
    lang_eng.append("English")
```

```python
print(data_it)
print(lang_it)
```

```python
final_data = pd.DataFrame({"Text":data_eng+data_ger+data_fre+data_spa+data_dutch+data_danish+data_it, "Language":lang_eng+lang_ger+lang_fre+lang_spa+lang_dutch+lang_danish+lang_it})
```

```python
print(final_data.shape)
```

```python
final_data.head()
```

```python
x ,y = final_data.iloc[:,0],final_data.iloc[:,1]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
vectorizer = feature_extraction.text.TfidfVectorizer(ngram_range=(1,3),analyzer='char')

pipe_lr_r13 = pipeline.Pipeline([
                                 ('vectorizer', vectorizer),
                                 ('clf', linear_model.LogisticRegression())
])

pipe_lr_r13.fit(X_train,y_train)
```

```python
y_predicted = pipe_lr_r13.predict(X_test)

acc = (metrics.accuracy_score(y_test, y_predicted))*100
print(acc,'%')
```

```python
matrix = metrics.confusion_matrix(y_test, y_predicted)
print('Confusion matrix: \n', matrix)
```
